```python
# Company Revenue Relationship 
# Numerize Company Data 

df_numerized = df

for col_name in df_numerized.columns: 
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
    
```

```python
# Now has numeric representation of each non-numeric value
correlation_matrix = df_numerized.corr(method='spearman')
sns.heatmap(correlation_matrix, annot = True)
plt.title('Corrilation Heatmap using Spearman Method')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show
```

```python
# Using Unstacking 
corrilation_mat = df_numerized.corr()
corr_pairs = corrilation_mat.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs.head()
```

```python
#Skimming for high corrliation elements 
High_Corrilations = sorted_pairs[(sorted_pairs) >  .5]
High_Corrilations.head()
```
